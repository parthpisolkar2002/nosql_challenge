5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
# Retrieve all documents that need updating
documents = establishments.find()

# Loop through each document
for document in documents:
    # Convert the longitude and latitude fields to float if they exist
    if 'longitude' in document and 'latitude' in document:
        try:
            longitude = float(document['longitude'])
            latitude = float(document['latitude'])
            
            # Update the document with the new data types
            establishments.update_one(
                {'_id': document['_id']},
                {'$set': {'longitude': longitude, 'latitude': latitude}}
            )
            print(f"Updated document ID: {document['_id']}")
        except Exception as e:
            print(f"Could not update document ID: {document['_id']}. Error: {e}")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```
